```python
import pandas as pd
```

```python
movies=pd.read_csv('top10K-TMDB-movies.csv')
```

```python
movies.head(5)
```

```python
movies=movies[['id', 'title', 'overview', 'genre']]
```

```python
movies
```

```python
movies['tags'] = movies['overview']+movies['genre']
```

```python
movies
```

```python


new_data  = movies.drop(columns=['overview', 'genre'])
```

```python
new_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=10000, stop_words='english')
cv
```

```python
vector=cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
vector.shape  #inculding the new data use the text and then need the valuse and convert the types
        #in the unoques vector and after the convert the array
```

```python
from sklearn.metrics.pairwise import cosine_similarity #2 action movie simalirt the text it call 
similarity=cosine_similarity(vector)
similarity  #
```

```python
new_data[new_data['title']=="The Godfather"].index[0] #check the index number the new data set
```

```python
distance = sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vector:vector[1])
for i in distance[0:5]:  #to the simlrirty of the each movies
    print(new_data.iloc[i[0]].title)
```

```python
def recommand(movies):    
    index=new_data[new_data['title']==movies].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
    for i in distance[0:5]:
        print(new_data.iloc[i[0]].title)
```

```python
recommand("Iron Man")  #recommed teh iron men movie
```
